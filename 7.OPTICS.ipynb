```python
from sklearn.cluster import OPTICS 
optics=OPTICS (min_samples=8,metric='minkowski')
Clustering=optics.fit_predict(X)
```

```python
Clustering
```

```python
from sklearn.metrics import pairwise_distances
met=pairwise_distances(X,metric='cosine')
```

```python
met
```

```python
from sklearn.metrics.pairwise import pairwise_kernels
ker=pairwise_kernels(X,metric='linear')
```

```python
ker
```

```python
from sklearn.metrics.pairwise import distance_metrics #valid distance metrics (calculate pair_wise)
metrics=distance_metrics()
```

```python
metrics
```

```python
from scipy.spatial.distance import pdist
pdist(X,metric='braycurtis')
```
